```python
import pandas as pd
import re
import random
```

```python
d = pd.read_csv('./clothes2.csv',header= 0)
df = pd.read_csv('./clothes.csv')
# 合并
df = pd.concat([d, df], ignore_index=True)
```

```python
# 根据商品id去重
df.drop_duplicates(subset=['商品ID'], inplace=True)
```

```python
# 删除价格字段为空和不是数字的行
df.dropna(subset=['价格'], inplace=True)
```

```python
# 把原价描述字段为NAN填充为券后价
df['原价描述'] = df['原价描述'].fillna("券后价")
```

```python
def parse_sale_volume(x):
    if pd.isna(x):
        return None
    x = str(x)
    x = x.replace("人付款", "").replace("人看过", "").strip()
    if "万+" in x:
        num = x.replace("万+", "")
        base = int(float(num) * 10000)
        return random.randint(base, base + 3000)
    if "+" in x:
        base = int(re.findall(r"\d+", x)[0])
        return random.randint(base, base + int(base * 0.2))
    return int(re.findall(r"\d+", x)[0])

df["实际销量"] = df["实际销量"].apply(parse_sale_volume)
```

```python
df["是否广告"] = df["是否广告"].astype(int)
```

```python
# 处理地区
def split_location(x):
    if pd.isna(x):
        return pd.Series([None, None])
    x = str(x).strip()
    parts = x.split()
    if len(parts) == 1:
        return pd.Series([parts[0], parts[0]])
    else:
        return pd.Series([parts[0], parts[1]])

df[["province", "city"]] = df["发货地"].apply(split_location)
```

```python
# 清洗出版型和面料
def extract_style_material(x):
    # 默认值
    style = "未知"
    material = "未知"

    if pd.isna(x):
        return pd.Series([style, material])

    x = str(x)

    # 匹配 版型
    style_match = re.search(r"版型:([^,'\]]+)", x)
    if style_match:
        style = style_match.group(1)

    # 匹配 面料
    material_match = re.search(r"面料:([^,'\]]+)", x)
    if material_match:
        material = material_match.group(1)

    return pd.Series([style, material])
df[["版型", "面料"]] = df["属性"].apply(extract_style_material)
```

```python
# 转换热门标签变成列表
import ast
def clean_hot_tags(x):
    if pd.isna(x):
        return []
    if isinstance(x, dict):
        if "rank_short_text" in x and x["rank_short_text"]:
            return [x["rank_short_text"]]
        else:
            return []
    if isinstance(x, str):
        x = x.strip()

        if x == "" or x == "{}":
            return []

        try:
            d = ast.literal_eval(x)  # 安全转 dict
            if isinstance(d, dict) and "rank_short_text" in d and d["rank_short_text"]:
                return [d["rank_short_text"]]
            else:
                return []
        except:
            return []
    return []
df["热门标签"] = df["热门标签"].apply(clean_hot_tags)
```

```python
# 删除字段
df.drop(columns=["关键词", "页码"], inplace=True)
```

```python
import numpy as np
n = len(df)
# 添加随机时间字段格式为2025-12-05 17:28:35 时间区段2025-01-01 00:00:00 - 2025-12-5 23:59:59
# 时间范围
start = pd.Timestamp("2025-01-01 00:00:00").timestamp()
end   = pd.Timestamp("2025-12-05 23:59:59").timestamp()
# 生成随机时间戳
random_timestamps = np.random.uniform(start, end, n)
# 转换为 datetime
df["时间"] = pd.to_datetime(random_timestamps, unit='s')
```

```python
# 存入数据库
df = df.rename(columns={
    "商品ID": "product_id",
    "标题": "title",
    "价格": "price",
    "原价描述": "original_price_desc",
    "店铺名称": "shop_name",
    "实际销量": "real_sales",
    "发货地": "ship_from",
    "是否广告": "is_ad",
    "店铺URL": "shop_url",
    "商品链接": "product_url",
    "优惠标签": "coupon_tag",
    "属性": "attributes",
    "商品封面": "cover_img",
    "热门标签": "hot_tag",
    "商品卖点": "selling_point",
    "店铺标签": "shop_tag",
    "province": "province",
    "city": "city",
    "版型": "style_type",
    "面料": "material",
    "时间": "create_time"
})
df["real_sales"] = df["real_sales"].fillna(0).astype(int)

df["is_ad"] = df["is_ad"].astype(int)
text_cols = [
    "title", "original_price_desc", "shop_name", "ship_from",
    "shop_url", "product_url", "coupon_tag", "attributes",
    "cover_img", "hot_tag", "selling_point", "shop_tag",
    "province", "city", "style_type", "material"
]
for col in text_cols:
    df[col] = df[col].fillna("").astype(str)
```
